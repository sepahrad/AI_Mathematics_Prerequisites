```python
from scipy.optimize import minimize
```

```python
def f(x):
  return x ** 2

def g(x):
  return  x - 5
```

```python
res = minimize(f, [0], constraints={'type': 'eq', 'fun': g})
print(res.x)
print(res.fun)
```

```python
def f(A):
  x, y = A
  return x ** 2 + y ** 2

def g(A):
  x, y = A
  return  x + y - 1
```

```python
res = minimize(f, [0, 0], constraints={'type': 'eq', 'fun': g})
print(res.x)
print(res.fun)
```

```python
def f(A):
  x, y = A
  return x ** 2 + y ** 2

def g1(A):
  x, y = A
  return  x + 1

def g2(A):
  x, y = A
  return  y + 1
```

```python
const = ({'type': 'eq', 'fun': g1},
         {'type': 'eq', 'fun': g2})

res = minimize(f, [0, 0], constraints=const)
print(res.x)
print(res.fun)
```
