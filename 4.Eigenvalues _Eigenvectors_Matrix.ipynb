We should find the value of L and V that satisfy the this equation:
A . V == L * V

---



```python
import numpy as np
```

```python
A = np.array([[1,4],[3,2]])
print(A)
```

```python
L, V = np.linalg.eig(A)
```

```python
L # Eigenvalues
```

```python
print(np.round(V, decimals=1)) #Eigenvectors
```

```python
L[0] * V[:, 0]
```

```python
np.dot(A, V[:, 0])
```

```python
L[1] * V[:, 1]
```

```python
np.dot(A, V[:, 1])
```
