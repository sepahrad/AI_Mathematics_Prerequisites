```python
x = [1, 2, 3]
y = [12, 14, 16]
n = len(x)

xbar = np.mean(x)
xbar
```

```python
ybar = np.mean(y)
ybar
```

```python
lst = []
for i in range(n):
    lst.append((x[i] - xbar)*(y[i] - ybar))

cov = sum(lst) / n
lst
print(cov)
```

```python
x = [1, 2, 3]
y = [12, 14, 16]
```

```python
sx = np.var(x) ** (1/2)
sy = np.var(y) ** (1/2)
cov = np.cov(x, y, ddof=0)[0][1]
p = cov / (sx * sy)
print(p)
```

```python
r, p = scipy.stats.pearsonr(x, y)
print(r)
```
